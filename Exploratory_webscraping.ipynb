```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
#url = "https://internshala.com/jobs/work-from-home"
```

```python
url = "https://internshala.com/jobs/"
```

```python
a = requests.get(url)
```

```python
soup = BeautifulSoup(a.text,"lxml")
soup
```

```python
jobs = soup.find_all("div", class_ = "internship_meta")
```

```python
jobs
```

```python
job_role = []
company_name = []
more_info = []
work_place = []
start_date = []
```

```python
for job in jobs:
  job_role.append(job.find("div", class_ = "heading_4_5 profile").text.strip())
  company_name.append(job.find("div", class_ = "heading_6 company_name").text.strip())
  logo = job.find("div", class_ = "internship_logo").text
  more_info.append(job.a["href"])
  work_place.append(job.find("p", {"id": "location_names"}).text.strip())
  salary = job.find("div",class_= "other_detail_item_row").text
  start_date.append(job.find("div", class_ = "item_body").text.strip())






```

```python
scraped_data = {'job_role':job_role,'company_name': company_name, 'more_info':more_info,'work_place':work_place,'start_date':start_date}
```

```python
job_data = pd.DataFrame(scraped_data)
```

```python
job_data
```

```python
job_data.to_csv("job_data.csv")
```
