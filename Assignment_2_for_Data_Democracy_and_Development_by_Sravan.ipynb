```python
wage = pd.read_csv('wage.csv')
wage.info()
```

```python
wage.head()
```

```python
#Solving for 3a
from statsmodels.formula.api import ols
ols('wage~educ', data = wage).fit().summary()
```

Here, t-statistic is greater than 1.96 which means the co-efficient of education is statistically significantly different from zero.

R_squared = 0.165
