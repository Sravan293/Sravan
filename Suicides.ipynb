```python
import pandas as pd
df = pd.read_csv('crimedata_suicides.csv')
```

```python
df['STATE.UT'].unique()
```

```python
tn_t = df[df['STATE.UT']=='TAMIL NADU']
tn = tn_t[tn_t['Year']!=2014]
tn.shape
```

```python
a=tn[tn['CAUSE']=='Total']

```

```python
import matplotlib.pyplot as plt

x = a['Year']
y = a['Total.Male']
y1 = a['Total.Female']
plt.plot(x,y,label ='Male')
plt.plot(x,y1,label ='Female')
plt.title('Farmer suicides')
plt.legend()
plt.show()
```

###### All India average in different categories among male and female compared to Tamilnadu.
###### Farmer suicides over the years
###### Trans Gender data issue
###### Employed, Unemployed, Self-employed
###### Housewifes among female
###### Age distribution

```python
#a=tn[(tn['CAUSE']=='Total')| (tn['CAUSE']=='Student')|(tn['CAUSE']=='Total Salaried')|(tn['CAUSE']=='Total Self-employed')|(tn['CAUSE']=='Unemployed')]

a0 = df[df['Year']!=2014]
```

```python
# All India average
a1=a0.groupby('CAUSE')
a2= (a1['Grand.Total'].sum())/37
national_avg=[]
for i in a2:
  national_avg.append(int(i))
```

```python
# Tamil Nadu average
b1 = tn.groupby('CAUSE')
b2 = b1['Grand.Total'].sum()
tn_avg =[]

for i in b2:
  tn_avg.append(int(i))
```

```python
# All India Male average
a1=a0.groupby('CAUSE')
m= (a1['Total.Male'].sum())/37
national_m_avg=[]
for i in m:
  national_m_avg.append(int(i))
```

```python
# All India Female Average
a1=a0.groupby('CAUSE')
f= (a1['Total.Female'].sum())/37
national_f_avg=[]
for i in f:
  national_f_avg.append(int(i))
```

```python
# Tamil nadu Male Average
b1 = tn.groupby('CAUSE')
m = b1['Total.Male'].sum()
tn_m_avg =[]

for i in m:
  tn_m_avg.append(int(i))
```

```python
b1 = tn.groupby('CAUSE')
f = b1['Total.Female'].sum()
tn_f_avg =[]

for i in f:
  tn_f_avg.append(int(i))
```

```python
d= tn['CAUSE'].unique()
e =[]
for i in d:
  e.append(i)
```

```python
e
```

```python
from tabulate import tabulate
print(tabulate({'CAUSE': e,
                'India':national_avg,
                'India Male': national_m_avg,
                'India Female':national_f_avg,
                'Tamil Nadu': tn_avg,
                'Tamil Nadu Male':tn_m_avg,
                'Tamil Nadu Female':tn_f_avg},
                headers='keys',
                tablefmt='fancy_grid',
                missingval='N/A'))
```

```python
new_df = pd.DataFrame({'CAUSE': e,'India':national_avg, 'India Male': national_m_avg,'India Female':national_f_avg,'Tamil Nadu': tn_avg,'Tamil Nadu Male':tn_m_avg,'Tamil Nadu Female':tn_f_avg})
```

```python
new_df.to_csv('table.csv')
```
